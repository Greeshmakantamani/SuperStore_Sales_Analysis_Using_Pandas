```python
new_df["Postal Code"] = new_df["Postal Code"].astype(str)
```

```python
new_df["Order Date"] = pd.to_datetime(new_df["Order Date"], format="mixed")
new_df["Ship Date"] = pd.to_datetime(new_df["Ship Date"], format="mixed")
```

```python
new_df["Sales"]=new_df["Sales"]*100/new_df["Sales"].sum()
new_df[["Country","Sales"]]
```

```python
new_df["Sales"].sum()
```

```python
new_df["Profit_percentages"]=new_df["Profit"] / new_df["Sales"] * 100     #Create a new column called "Profit_Percentage" using:

#(Profit / Sales) * 100
```

```python
new_df.columns   # Profit_Percentages is added
```

```python
new_df[["Country","Sales","Profit","Profit_percentages"]]
```

```python
new_df[(new_df["Profit"] >0) | (new_df["Sales"] >50)]
```

```python
new_df.groupby("Country")["Sales"].sum()
new_df.groupby("Country")["Profit"].sum()
new_df.groupby("Country")["Profit_percentages"].mean()
```
